```python
import sys
import os
import pandas as pd
import joblib

# Añadimos la raíz para importar src
sys.path.append('..')
from src.data_preprocessing import load_gym_data
from src.model_training import get_regression_pipeline, get_classification_pipeline

# Configurar rutas para guardar los modelos
MODELS_DIR = '../models/trained_models/'
os.makedirs(MODELS_DIR, exist_ok=True)

# 1. Cargar Datos
df = load_gym_data('../data/gym_members_exercise_tracking.csv')

# ==========================================
# TAREA 1: REGRESIÓN (Predecir Fat_Percentage)
# ==========================================
# Justificación: Estimar el % de grasa con variables de entrenamiento sin usar caliper.
X_reg = df.drop(columns=['Fat_Percentage'])
y_reg = df['Fat_Percentage']

# Instanciar pipelines
lin_reg_pipe = get_regression_pipeline(model_type='linear')
rf_reg_pipe = get_regression_pipeline(model_type='random_forest')

# Entrenar
print("Entrenando modelos de regresión...")
lin_reg_pipe.fit(X_reg, y_reg)
rf_reg_pipe.fit(X_reg, y_reg)

# Serializar y guardar
joblib.dump(lin_reg_pipe, os.path.join(MODELS_DIR, 'linear_regression_model.pkl'))
joblib.dump(rf_reg_pipe, os.path.join(MODELS_DIR, 'random_forest_regressor_model.pkl'))
print("Modelos de regresión guardados en models/trained_models/")

# ==========================================
# TAREA 2: CLASIFICACIÓN (Predecir Experience_Level)
# ==========================================
# Justificación: Identificar el nivel del atleta para automatizar recomendaciones de rutinas.
X_clf = df.drop(columns=['Experience_Level'])
y_clf = df['Experience_Level']

# Instanciar pipelines
log_reg_pipe = get_classification_pipeline(model_type='logistic')
rf_clf_pipe = get_classification_pipeline(model_type='random_forest')

# Entrenar
print("Entrenando modelos de clasificación...")
log_reg_pipe.fit(X_clf, y_clf)
rf_clf_pipe.fit(X_clf, y_clf)

# Serializar y guardar
joblib.dump(log_reg_pipe, os.path.join(MODELS_DIR, 'logistic_regression_model.pkl'))
joblib.dump(rf_clf_pipe, os.path.join(MODELS_DIR, 'random_forest_classifier_model.pkl'))
print("Modelos de clasificación guardados en models/trained_models/")
```
