### Elección de Algoritmos y Prevención de Data Leakage
Se implementa un **Pipeline** que integra el preprocesamiento y el modelo. Esto es crítico para asegurar la **reproducibilidad total** y evitar la fuga de información (Data Leakage) durante la validación cruzada, garantizando que el escalado de datos se realice solo con información del conjunto de entrenamiento.

Se elige **Random Forest** y **KNN** sobre modelos lineales simples debido a que las variables fisiológicas (como la relación entre IMC, edad y gasto calórico) presentan relaciones no lineales complejas que requieren modelos con mayor capacidad de captura de patrones.

```python
import sys
import os
import pandas as pd
import joblib

# Añadimos la raíz para importar src
sys.path.append('..')
from src.data_preprocessing import load_gym_data
from src.model_training import get_regression_pipeline, get_classification_pipeline

# Configurar rutas para guardar los modelos
MODELS_DIR = '../models/trained_models/'
os.makedirs(MODELS_DIR, exist_ok=True)

# 1. Cargar Datos
df = load_gym_data('../data/gym_members_exercise_tracking.csv')

# ==========================================
# TAREA 1: REGRESIÓN (Predecir Fat_Percentage)
# ==========================================
X_reg = df.drop(columns=['Fat_Percentage'])
y_reg = df['Fat_Percentage']

# Instanciar pipelines (Incluyendo KNN)
lin_reg_pipe = get_regression_pipeline(model_type='linear')
rf_reg_pipe = get_regression_pipeline(model_type='random_forest')
knn_reg_euc_pipe = get_regression_pipeline(model_type='knn', knn_metric='euclidean')
knn_reg_man_pipe = get_regression_pipeline(model_type='knn', knn_metric='manhattan')

# Entrenar
print("Entrenando modelos de regresión...")
lin_reg_pipe.fit(X_reg, y_reg)
rf_reg_pipe.fit(X_reg, y_reg)
knn_reg_euc_pipe.fit(X_reg, y_reg)
knn_reg_man_pipe.fit(X_reg, y_reg)

# Serializar y guardar
joblib.dump(lin_reg_pipe, os.path.join(MODELS_DIR, 'linear_regression_model.pkl'))
joblib.dump(rf_reg_pipe, os.path.join(MODELS_DIR, 'random_forest_regressor_model.pkl'))
joblib.dump(knn_reg_euc_pipe, os.path.join(MODELS_DIR, 'knn_regression_euclidean.pkl'))
joblib.dump(knn_reg_man_pipe, os.path.join(MODELS_DIR, 'knn_regression_manhattan.pkl'))
print("Modelos de regresión (incluyendo KNN) guardados en models/trained_models/")

# ==========================================
# TAREA 2: CLASIFICACIÓN (Predecir Experience_Level)
# ==========================================
X_clf = df.drop(columns=['Experience_Level'])
y_clf = df['Experience_Level']

# Instanciar pipelines (Incluyendo KNN)
log_reg_pipe = get_classification_pipeline(model_type='logistic')
rf_clf_pipe = get_classification_pipeline(model_type='random_forest')
knn_clf_euc_pipe = get_classification_pipeline(model_type='knn', knn_metric='euclidean')
knn_clf_man_pipe = get_classification_pipeline(model_type='knn', knn_metric='manhattan')

# Entrenar
print("\nEntrenando modelos de clasificación...")
log_reg_pipe.fit(X_clf, y_clf)
rf_clf_pipe.fit(X_clf, y_clf)
knn_clf_euc_pipe.fit(X_clf, y_clf)
knn_clf_man_pipe.fit(X_clf, y_clf)

# Serializar y guardar
joblib.dump(log_reg_pipe, os.path.join(MODELS_DIR, 'logistic_regression_model.pkl'))
joblib.dump(rf_clf_pipe, os.path.join(MODELS_DIR, 'random_forest_classifier_model.pkl'))
joblib.dump(knn_clf_euc_pipe, os.path.join(MODELS_DIR, 'knn_classification_euclidean.pkl'))
joblib.dump(knn_clf_man_pipe, os.path.join(MODELS_DIR, 'knn_classification_manhattan.pkl'))
print("Modelos de clasificación (incluyendo KNN) guardados en models/trained_models/")
```
